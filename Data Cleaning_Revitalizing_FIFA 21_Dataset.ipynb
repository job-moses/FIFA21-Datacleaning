### 4.4 Transforming 'Contract' column into more usable format

**A close look at the columns we could notice that the columns provide informartion on**
- players that their contract is active

**let define a function  to create a new feature from this column**

```python
#define function to create new feature from contract column
def contract_status(x):
    if '~' in x:
        x = 'Active'
        return x
    elif 'Loan' in x:
        x = 'On Loan'
        return x
    elif 'Free' in x:
        x = 'Free'
        return x
```

### 4.5 Transforming player  "Height" and "Weight"  columns

```python
#check for unique values
for cols in df[["Height", "Weight" ]]:
    print(cols)
    print (df[cols].unique())
    print('-' * 50)
```

#### The height consist of two measurement **Centimeters** and **feet with inches**
- let define a function to transform the feet and inches to equivalent values in centimeters
- Strip  "cm"
- convert the values to integer data type
#### The Weight consist of two measurement **kilogram(kg)** and **pound(lbs)**
- let define a function to transform the pound(lbs) to equivalent values in Kg
- Strip  "Kg"
- convert the values to integer data type

```python
# Define a fuction to convert values in feet and inches in Height to equivalent value in centimeters
def all_to_cm(x):
    if 'cm' in x:
        x = x.strip('cm')
        return int(x)
    elif '"' and "'" in x:
        feet,inches = x.split("'")
        total_height = int(feet) * 12 +  int(inches.strip('"'))
        x= round(total_height * 2.54)
        return x
```

```python
# fuction to convert lbs to kg in weight column
def lbs_to_kg(x):
    if 'kg' in x:
        x = x.strip('kg')
        return int(x)
    elif 'lbs' in x:
        height = x.strip('lbs')
        x = int(height)  * 0.45359237
        return round(x)
```

**.apply()**

```python
df['Weight_kg'] = df['Weight'].apply(lbs_to_kg).astype('int')
df['Height_cm'] = df['Height'].apply(all_to_cm).astype('int')
```

```python
#check for unique values
for cols in df[["Height_cm", "Weight_kg" ]]:
    print(cols)
    print (df[cols].unique())
    print (df[cols].dtype)
    print('-' * 50)
```

### 4.6 Transforming player  "Wage" , "Value" , "Relase Clause" columns

```python
# check for unique value
for cols in df[['Wage','Value', 'Release Clause']]:
    print(cols)
    print (df[cols].unique())
    print('-' * 50)
```

**Define function to clean columns values**

```python
# define function to clean wage
def wage(x):
    if 'K' in x:
        wage = x.strip('€K')
        x = int(wage) * 1000
        return x
    else:
        x = x.strip('€')
        return int(x)
```

```python
# function to clean 'Value' and 'Release Clause'
def value(x):
    if 'M' in x:
        value = x.strip('€M')
        x = float(value) * 1000000
        return x
    elif "K" in x:
        value = x.strip('€K')
        x = float(value) * 1000
        return x
    else:
        x = x.strip('€')
        return float(x)
```

**.apply()**

```python
## Apply fuction to clean columns
#apply fuction to clean the column
df['Wage_euro'] = df['Wage'].apply(wage)
df['Value_euro'] = df['Value'].apply(value)
df[ 'Release Clause_euro'] = df[ 'Release Clause'].apply(value)
```

```python
# inspect changes
for cols in df[['Wage_euro','Value_euro', 'Release Clause_euro']]:
    print(cols)
    print (df[cols].unique())
    print('-' * 50)
```

### 4.9 Transforming the "Hits" column

```python
df['Hits'].unique()
```

```python
#extract numeric value unit into a new column
df['player hits'] = df['Hits'].str.extract('(.+[^K])').astype('float')

#extract 'K' from values
df['player_hit_unit'] = df['Hits'].str.extract('(K)')

#converting the K to equivalent thousand value
df['player hits'] = df.loc[df['player_hit_unit'] == 'K','player hits'] = df['player hits']  * 1000

#fill null value in player_hits with zero
df['player hits'].fillna(0, inplace = True)
```
